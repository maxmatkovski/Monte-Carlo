```python
import random
import math
import matplotlib.pyplot as plt

def monte_carlo_pi(num_samples):
    """
    Estimate the value of pi using Monte Carlo simulation.
    """
    num_inside = 0
    for i in range(num_samples):
        # Generate a random point in the unit square
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        # Check if the point is inside the unit circle
        if math.sqrt(x**2 + y**2) <= 1:
            num_inside += 1
    # Estimate the value of pi based on the number of points inside the circle
    return 4 * num_inside / num_samples

# Run the simulation and plot the results
num_simulations = 100
num_samples = 1000
pi_estimates = [monte_carlo_pi(num_samples) for i in range(num_simulations)]
plt.hist(pi_estimates, bins=20)
plt.axvline(x=math.pi, color='r', linestyle='--')
plt.xlabel('Value of pi')
plt.ylabel('Frequency')
plt.title(f'Monte Carlo Estimation of Pi\n{num_simulations} simulations, {num_samples} samples each')
plt.show()
```
